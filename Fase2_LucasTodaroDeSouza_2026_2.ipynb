Como os arquivos criptografados guardavam informações importantes, Miguel pensou em uma ideia para comprimir os arquivos sem perder informações.

Miguel sabe que os arquivos estão criptografados utilizando apenas as letras ***a, b, c*** e que, a criptografia possui as seguintes regras:


* Sequência Obrigatória: O texto deve ser composto exclusivamente pelos caracteres 'a', 'b' e 'c', nesta ordem exata. Porém, não é garantida a existência dos 3 caracteres em um texto.

* Início Fixo: O texto deve obrigatoriamente começar com o caractere 'a'.

* Transições Únicas: Uma vez que a sequência mudou de 'a' para 'b', nenhum 'a' pode aparecer novamente. O mesmo vale para a transição de 'b' para 'c'.

* Dependência: O texto só pode ter caracteres 'b' se existirem caracteres 'a', e só pode ter caracteres 'c' se existirem caracteres 'b'.

* Tamanho: É garantido que os arquivos nunca estejam vazios

* Conclusão: Após o caractere 'c' aparecer, nenhum 'a' ou 'b' é permitido.

Porém, sabe-se que alguns arquivos foram danificados durante a criptografia, e tiveram alguns caracteres trocaram de posições, ou até mesmo novos caracteres, ou então, até mesmo novos caracteres diferentes de 'a','b','c' foram inseridos! **Os arquivos danificados NÃO PODERÃO ser comprimidos. Mas devem ser identificados.**

Sabendo disso, Miguel decidiu **comprimir o conteúdo do arquivo para o formato "aXbYcZ"**, onde X, Y, Z são as quantidades dos caracteres 'a', 'b', 'c', respectivamente.

### **Tarefa:**

Seu objetivo é criar um programa em Python que receba o texto contido em um arquivo da empresa de Miguel, verifique se ele é válido, e comprima o texto do arquivo para o novo formato indicado por Miguel.

### **Entrada:**

A sua função deverá receber como entrada, uma ***string S***, que representará um texto criptografado válido nos arquivos da empresa de Miguel. A string de entrada **não obrigatoriamente estará no formato válido!**

### **Saída:**

Seu programa deverá retornar a String de entrada para a sua versão compactada, caso a string fornecida na entrada for válida, caso contrário, se ela não for válida, seu programa deve retornar "NÃO É POSSIVEL COMPRIMIR"

````python
# ============================================================
# Setup — Imports e funções auxiliares
# ============================================================

from PIL import Image
import cv2
import requests
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO


def img_from_url(url):
    """
    Baixa uma imagem a partir de uma URL e retorna uma imagem PIL em RGB.
    """
    response = requests.get(url, stream=True)
    response.raise_for_status()
    img = Image.open(BytesIO(response.content)).convert("RGB")
    return img


def convert_from_image_to_cv2(img):
    """
    Converte uma imagem PIL RGB para o formato usado pelo OpenCV, isto é, BGR.
    """
    img_array = np.asarray(img)

    if len(img_array.shape) == 3 and img_array.shape[2] == 3:
        img_array = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

    return img_array


def convert_from_cv2_to_rgb(cv2_image):
    """
    Converte uma imagem OpenCV BGR para RGB, para exibição com matplotlib.
    """
    if len(cv2_image.shape) == 2:
        return cv2_image

    return cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)


def plot_side_by_side(img1, img2, title1="Original", title2="Processada", figsize=(12, 6)):
    """
    Exibe duas imagens lado a lado.
    """
    plt.figure(figsize=figsize)

    plt.subplot(1, 2, 1)
    if isinstance(img1, Image.Image):
        plt.imshow(img1)
    elif len(img1.shape) == 2:
        plt.imshow(img1, cmap="gray")
    else:
        plt.imshow(img1)
    plt.title(title1)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    if isinstance(img2, Image.Image):
        plt.imshow(img2)
    elif len(img2.shape) == 2:
        plt.imshow(img2, cmap="gray")
    else:
        plt.imshow(img2)
    plt.title(title2)
    plt.axis("off")

    plt.tight_layout()
    plt.show()


def plot_many(images, titles, cmaps=None, figsize=(18, 6)):
    """
    Exibe múltiplas imagens lado a lado.
    """
    if cmaps is None:
        cmaps = [None] * len(images)

    plt.figure(figsize=figsize)

    for i, (img, title, cmap) in enumerate(zip(images, titles, cmaps), start=1):
        plt.subplot(1, len(images), i)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis("off")

    plt.tight_layout()
    plt.show()
````

## PARTE 3.5 — Balanceamento das Classes

Antes de partir para a engenharia de features, é necessário avaliar se as classes do seu problema estão bem distribuídas. Datasets desbalanceados são um dos problemas mais comuns em Machine Learning aplicado, e ignorá-los compromete diretamente a capacidade do modelo de aprender a identificar a classe minoritária.

### Contexto

Observe a distribuição da variável `review_score` no dataset. Você provavelmente notará que avaliações positivas (notas 4 e 5) são significativamente mais frequentes do que avaliações negativas (notas 1, 2 e 3). Isso caracteriza um **desbalanceamento de classes**, e, se não tratado, o modelo tende a simplesmente ignorar a classe minoritária, acertando muito na maioria e errando sistematicamente na minoria.

### Estratégias disponíveis

Existem diferentes abordagens para lidar com esse problema. As principais são:

| Estratégia | Como funciona | Quando usar |
|---|---|---|
| **Undersampling** | Reduz amostras da classe majoritária até equilibrar | Quando o dataset é grande e você pode perder dados |
| **Oversampling (SMOTE)** | Gera amostras sintéticas da classe minoritária | Quando o dataset é menor e você quer preservar informação |
| **class_weight** | Penaliza mais os erros na classe minoritária durante o treino | Quando não quer alterar o dataset, apenas o comportamento do modelo |
